```python
#data cleaning - adjust N/A values
#I will substitute N/A values with avg of column - I will explain in the report why
```

```python
print(population.isna().sum())
print(poverty.isna().sum())
print(forest.isna().sum())
```

```python
population['tot2010'] = population['tot2010'].fillna(population['tot2010'].mean())
population['prop_urb2010'] = population['prop_urb2010'].fillna(population['prop_urb2010'].mean())
population['prop_rur2010'] = population['prop_rur2010'].fillna(population['prop_rur2010'].mean())
poverty['ipm'] = poverty['ipm'].fillna(poverty['ipm'].mean())
forest['deorg'] = forest['deorg'].fillna(forest['deorg'].mean())
```

```python
#check
print(population.isna().sum())
print(poverty.isna().sum())
print(forest.isna().sum())
```

```python
#now I need to work on the disease set to create a composite indicator of total disease burden (malaria, chagas, dengue and leishmaniasis) *ask Giulia for support
```

```python
disease = disease[disease['period'] == '2004-2008']

disease_burden = disease.groupby(['municipality', 'geocode'])['cases'].sum().reset_index()

disease_burden = disease_burden.rename(columns={'cases': 'Disease_Burden'})

print(disease_burden.head())
```

```python
disease_burden.to_csv('disease_burden.csv', index=False)
from google.colab import files
files.download('disease_burden.csv')
```

```python
#merge datasets - used Google Colab AI to know how to merge datasets *column names error *ask Giulia for support
```

```python
#there is a problem with columns; in the forest dataset and in the poverty one municipality is named municipality_code
forest = forest.rename(columns={'municipality_code': 'municipality'})
poverty = poverty.rename(columns={'municipality_code': 'municipality'})

Final_df = pd.merge(disease_burden, forest, on=['municipality', 'geocode'], how='left')
Final_df = pd.merge(Final_df, population, on=['municipality', 'geocode'], how='left')
Final_df = pd.merge(Final_df, poverty, on=['municipality', 'geocode'], how='left')
```

```python
#explore and clean the new merged data file
```

```python
print (Final_df.columns)
```

```python
selected_columns = ['municipality', 'geocode', 'Disease_Burden', 'deorg', 'tot2010', 'prop_urb2010', 'prop_rur2010', 'ipm']
Cleaned_Dataset = Final_df[selected_columns]
```

```python
print(Cleaned_Dataset.isna().sum())
```

```python
Cleaned_Dataset['ipm'] = Cleaned_Dataset['ipm'].fillna(poverty['ipm'].mean())
```

```python
print(Cleaned_Dataset.isna().sum())
```

```python
#when I downloaded the file for the first time, it had different duplicates; now I need to fix these
duplicates = Cleaned_Dataset[Cleaned_Dataset.duplicated(subset=['geocode', 'municipality'], keep=False)]
print(duplicates)
```

```python
Cleaned_Dataset = Cleaned_Dataset.groupby(['geocode', 'municipality'], as_index=False).mean()
```

```python
duplicates = Cleaned_Dataset[Cleaned_Dataset.duplicated(subset=['geocode', 'municipality'], keep=False)]
print(duplicates)
```
